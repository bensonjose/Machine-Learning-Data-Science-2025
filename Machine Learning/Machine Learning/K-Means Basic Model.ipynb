### K-Means Clustering

```python
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.cluster import KMeans  
```

```python
# Generating some random 2D data points (100 points total)
X = np.random.rand(100, 2)  # Values range between 0 and 1
```

```python
# Setting up K-Means with 3 clusters 
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)  
```

```python
# Training the model on the data (finding cluster centers)
kmeans.fit(X)  
```

```python
# Predicting which cluster each point belongs to
y_kmeans = kmeans.predict(X)  
```

```python
# Plotting the data points, color-coded by their clusters

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', edgecolor='k') 



# Now, marking the cluster centers distinctly

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, marker='X', c='red', label='Centers') 


plt.legend()  
plt.title("K-Means Clustering") 
plt.show()
```
