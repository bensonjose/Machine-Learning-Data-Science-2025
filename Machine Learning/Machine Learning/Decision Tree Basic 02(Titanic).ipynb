```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Dell\Downloads\titanic (1).csv")
df.head()
```

```python
df.shape
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')

target = df.Survived
```

```python
# Manually encoding 'Sex' as numbers (1=male, 2=female) instead of using LabelEncoder

inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
# Selects rows from index 0 to 9 in the 'Age' column

inputs.Age[:10] 
```

```python
# Replace missing values in 'Age' with the column's mean for consistency in model training

inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting dataset into training (80%) and testing (20%) sets using train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
# Checking the number of samples in the training and testing sets

len(X_train)  
```

```python
len(X_test) 
```

```python
df.shape
```

```python
# Creating a Decision Tree model using scikit-learn's tree module

from sklearn import tree

# Initializes a basic decision tree classifier
model = tree.DecisionTreeClassifier()  # Initializes a basic decision tree classifier
```

```python
# Training the Decision Tree model using the training dataset

model.fit(X_train, y_train)  # The model learns patterns from the training data
```

```python
# Evaluating the accuracy of the trained Decision Tree model on the test dataset
 
model.score(X_test, y_test)  # Returns accuracy score (between 0 and 1)
```
