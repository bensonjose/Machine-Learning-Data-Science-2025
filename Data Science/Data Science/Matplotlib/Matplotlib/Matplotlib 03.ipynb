## Exercise 3: Read all product sales data and show it using a multiline plot.

## Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv(r"C:\Users\Dell\Downloads\Data Science Files\Matplotlib Explanation\11_Python_Matplotlib_Module-main\company_sales_data.csv")

# Extract the profit data
profitList = df['total_profit'].tolist()

# Set the figure size
plt.figure(figsize=(8, 5))

# Define custom profit ranges (bins)
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]

# Create the histogram using the custom profit ranges as bins
plt.hist(profitList, bins=profit_range, edgecolor='black', label='Profit data')

# Add gridlines
plt.grid(True, linewidth=1, linestyle="--")

# Set the labels and title
plt.xlabel('Profit range in dollars')
plt.ylabel('Frequency of Months')  # This represents the number of months
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Distribution of Profit Data')

# Display the plot
plt.show()
```

## Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.`

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv(r"C:\Users\Dell\Downloads\Data Science Files\Matplotlib Explanation\11_Python_Matplotlib_Module-main\company_sales_data.csv")

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df.facecream.sum(), df.facewash.sum(), df.toothpaste.sum(), 
         df.bathingsoap.sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
explode = [0, 0.1, 0, 0, 0, 0]

plt.axis("equal")  # Ensure the aspect ratio of the pie chart is equal

plt.pie(salesData,explode, labels=labels, autopct='%1.2f%%')
plt.legend(loc='upper left')
plt.title('Sales data')


plt.show()
```

## Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\Dell\Downloads\Data Science Files\Matplotlib Explanation\11_Python_Matplotlib_Module-main\company_sales_data.csv")

monthList=df.month_number.tolist()
bathingsoap=df.bathingsoap.tolist()
facewash=df.facewash.tolist()

# Creating a figure and an array of subplots (2 rows, shared x-axis)
figure,axarr=plt.subplots(2,sharex=True) # 'f'/figure captures the Figure object, 'axarr' captures the array of Axes objects(any name can be given for f and axarr but by convention use the standard names.)

# We access these subplots using index, since 2 is given it will go till 1, hence we use [0] and [1]

axarr[0].plot(monthList,bathingsoap,label="Bathing Soap Data",color="red",marker='o',linewidth=3)
axarr[0].set_title("Sub-plot for bathing soap")  # Set specific title for the first subplot
axarr[0].set_ylabel('Sales units in numbers')   # Set specific y-axis label for the first subplot

axarr[1].plot(monthList,facewash,label="Facewash Data",marker='o',color="k",linewidth=3)
axarr[1].set_title("Sub-plot for facewash")   # Set specific title for the second subplot
axarr[1].set_ylabel('Sales units in numbers')  # Set specific y-axis label for the second subplot

plt.legend()
plt.xticks(monthList)
plt.xlabel("Months") # Setting common x-axis label for the entire figure
# plt.ylabel("Sales units in numbers")  Using this would create common ylabel, meaning only for one, to create distinctinve use set_ylabel as above
# plt.title("Sub-plots") Using this would create common title, meaning only for one, to create distinctinve use set_title as above

plt.show()
```

## Exercise Question 10: Read all product sales data and show it using the stack plot

```python
import pandas as pd
import matplotlib.pyplot as plt  

# Read the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\Dell\Downloads\Data Science Files\Matplotlib Explanation\11_Python_Matplotlib_Module-main\company_sales_data.csv")

# Extract data into lists
monthList = df['month_number'].tolist()
faceCremSalesData = df['facecream'].tolist()
faceWashSalesData = df['facewash'].tolist()
toothPasteSalesData = df['toothpaste'].tolist()
bathingsoapSalesData = df['bathingsoap'].tolist()
shampooSalesData = df['shampoo'].tolist()
moisturizerSalesData = df['moisturizer'].tolist()

# Plot individual lines with custom line widths for the legend
# plt.plot([], [], color='m', label='Face Cream', linewidth=5)
# plt.plot([], [], color='c', label='Face Wash', linewidth=5)
# plt.plot([], [], color='r', label='Tooth Paste', linewidth=5)
# plt.plot([], [], color='k', label='Bathing Soap', linewidth=5)
# plt.plot([], [], color='g', label='Shampoo', linewidth=5)
# plt.plot([], [], color='y', label='Moisturizer', linewidth=5)


# Create the stack plot with sales data and labels
plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              labels=['face Cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer'],
              colors=['m','c','r','k','g','y'])

# Set labels and title for the plot
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.title('All product sales data using stack plot')

# Add legend to the plot
plt.legend(loc='upper left')

# Display the plot
plt.show()
```
