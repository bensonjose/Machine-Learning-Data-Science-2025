**<h1>Line Plots</h1>**

**Plotting a Simple Line Graph**

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Plotting the data
plt.plot(x, y, label='Linear')

# Adding labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Plot')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

**Plotting a Simple Line Graph Using a Pandas DataFrame**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data in a DataFrame
data = {
    'X': [0, 1, 2, 3, 4],
    'Y': [0, 2, 4, 6, 8]
}
df = pd.DataFrame(data)

# Plotting the data
plt.plot(df['X'], df['Y'], label='Linear')

# Adding labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Plot with DataFrame')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

**Plotting Two Lines Using a Pandas DataFrame**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data in a DataFrame
data = {
    'X': [0, 1, 2, 3, 4],
    'Y1': [0, 2, 4, 6, 8],
    'Y2': [0, 1, 2, 3, 4]
}
df = pd.DataFrame(data)

plt.figure(figsize=(8,5), dpi=100)  # Use 300 dpi for better clarity.(when printing or saving images)

# Plotting the first line
plt.plot(df['X'], df['Y1'], label='Y1 Line', color='blue', marker='o',linestyle=":")

# Plotting the second line
plt.plot(df['X'], df['Y2'], label='Y2 Line', color='red', marker='s',linestyle='--')

# Adding labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Two Lines Plot with DataFrame')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

**<h1>Bar Chart</h1>**

```python
# Data for the bar chart
categories = ['A', 'B', 'C']
values = [1, 4, 2]

# Creating a bar chart
plt.bar(categories, values)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')

# Display the plot
plt.show()
```

```python

# Data for Bar Chart
categories = ['A', 'B', 'C']
values = [1, 4, 2]

# Creating a bar chart
bars = plt.bar(categories, values)

# Define the hatch patterns
patterns = ['/', 'O', '*']

# Apply hatch patterns to each bar using a loop
for i, bar in enumerate(bars):
    # print(i)  # Print the index
    bar.set_hatch(patterns[i])

# Labels
plt.xlabel('Categories')
plt.ylabel('Values')

#Title
plt.title('Bar Chart with Hatch Patterns')

# Saving
# plt.savefig('barchart.png', dpi=300)

# Display the plot
plt.show()
```

**Ticks**

**Ticks are the value on the axis to show the coordinates on the graph.**<br>
It is the value on the axes by which we can visualize where will a specific coordinate lie on a graph.

```python
import matplotlib.pyplot as plt

years = [2000, 2001, 2002, 2003, 2004, 2005]
values = [1, 3, 2, 5, 4, 6]

plt.plot(years, values)

# Customize x-axis ticks
plt.xticks([2000, 2002, 2004, 2005], labels=['Y2K', '2002', '2004', '05'])

plt.show()
```
