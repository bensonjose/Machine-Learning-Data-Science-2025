**Pandas Merge**

```python
import pandas as pd
```

```python
df1=pd.DataFrame({"City":["Seattle","Miami","Boston"],
                 "temperature":[10,19,23]})
df1
```

```python
df2=pd.DataFrame({"City":["Miami","Seattle","Boston","Detroit","Atlanta"],
                 "Humidity":[34,45,40,28,23]})
df2
```

```python
new_df1=pd.merge(df1,df2,on='City')
new_df1
```

```python
df3=pd.DataFrame({"City":["New York City","Los Angeles","Las Vegas","Boston","Seattle"],
                 "Temperature":[14,20,17,8,10]})
df3
```

```python
df4=pd.DataFrame({"City":["Los Angeles","New York City","Boston","Cincinnati","Seattle"],
                 "Humidity":[56,28,30,43,35]})
df4
```

```python
new_df2=pd.merge(df3,df4,how='inner')
new_df2
```

```python
new_df2=pd.merge(df3,df4,how='left')
new_df2
```

```python
new_df2=pd.merge(df3,df4,how='right')
new_df2
```

```python
new_df2=pd.merge(df3,df4,how='outer')
new_df2
```

```python
new_df2=pd.merge(df3,df4,how='outer',indicator=True)
new_df2
```

```python
df5 = pd.DataFrame({
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
    "temperature": [29, 24, 15, 32, 35],
    "Humidity": [60, 50, 55, 70, 20]
})
df5
```

```python
# Creating df6
df6 = pd.DataFrame({
    "City": ["Chicago", "Houston", "Phoenix", "San Antonio", "San Diego"],
    "temperature": [18, 30, 33, 28, 22],
    "Humidity": [58, 65, 25, 75, 55]
})
df6
```

```python
new_df3=pd.merge(df5,df6,on='City',how='outer')  #suffixes=('_first','_second')  to have your own naming
new_df3
```
