`**<h1>Data Visualization using Pandas</h1>**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
# This will apply the FiveThirtyEight style to your graphs.
# It's basically a style that gives your plots a modern and clean look.
```

```python
data=pd.Series(np.random.rand(1000).cumsum())
# This creates a pandas Series with the cumulative sum of 1000 random numbers ranging from 0 to 1
data
```

```python
data.plot()
```

```python
data1=pd.Series(np.random.randn(1000).cumsum())
# This creates a pandas Series with the cumulative sum of 1000 random numbers generated from a standard normal distributions
data1
```

```python
data1.plot()
```

```python
df=pd.DataFrame(np.random.randn(100,4).cumsum(axis=0),columns=['A','B','C','D'])
# Creating a DataFrame with the cumulative sum of random numbers in 100 rows and 4 columns
# Columns are labeled 'A', 'B', 'C', and 'D' . We mention axis=0 since this is s dataframe
df
```

```python
df.plot()
```

**<h1>Histograms</h1>**

```python
iris=pd.read_csv(r"C:\Users\Dell\Downloads\Data Science Files\Pandas Explanation\Pandas main\DataSets\iris.data", header=None)
iris.columns=["sepal_length","sepal_width", "petal_length", 
              "petal_width", "species"]
```

```python
iris.head()
```

```python
iris.plot.hist(alpha=0.7)
# This will plot histograms for each numerical column in the 'iris' DataFrame with slight transparency (alpha=0.7)
```

```python
iris.plot.hist(alpha=1,stacked=True)
```

```python
iris.plot.hist(alpha=1, stacked=True, bins=25)

# This will plot histograms for each numerical column in the 'iris' DataFrame with full opacity (alpha=1)
# The histograms will be stacked on top of each other and divided into 25 bins (intervals)
```

```python
iris["sepal_width"].plot.hist(orientation="horizontal")
# This will plot a histogram for the 'sepal_width' column in the 'iris' DataFrame with horizontal bars
```

```python
iris["sepal_length"].diff().hist()
```

```python
iris.hist(color="blue", alpha=1, bins=20)
# This will plot histograms for each numerical column in the 'iris' DataFrame
# The bars will be colored blue with full opacity (alpha=1)
# The histograms will be divided into 20 bins (intervals)
```

```python
iris.hist("petal_length",by="species")
# This will create histograms for the 'petal_length' column in the 'iris' DataFrame
# The histograms will be grouped by the 'species' column
```

**<h1>Boxplot charts</h1>**

```python
iris.plot.box() 
```

```python
colors={'boxes': 'Red', 'whiskers': 'blue','medians': 'Black', 'caps': 'Green'}
# This dictionary sets the colors for different elements of a box plot
# 'boxes': the main box, 'whiskers': the lines extending from the box, 'medians': the central line inside the box, 'caps': the end lines of the whiskers.
```

```python
iris.plot.box(color=colors)
```

```python
iris.plot.box(vert=False)
```

```python
iris.boxplot()
```

```python
# Setting the default figure size for plots to 8x8 inches
plt.rcParams["figure.figsize"]=(8,8)

# Using the 'ggplot' style for the plots
plt.style.use("ggplot")

# Creating a boxplot for the 'iris' DataFrame, grouping by the 'species' column
iris.boxplot(by='species')
```
