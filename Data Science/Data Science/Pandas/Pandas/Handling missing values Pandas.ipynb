**Handling Missing Data - fillna, interpolate, dropna**

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Dell\Downloads\weather_data (1).csv")
df
```

```python
df=pd.read_csv(r"C:\Users\Dell\Downloads\weather_data (1).csv",parse_dates=['day'])
type(df.day[0])
df
```

```python
df.set_index('day',inplace=True)
df
```

**Fillna
Fill all NaN with one specific value**

```python
new_df=df.fillna(0)
new_df
```

**Fill na using column names and dict**

```python
new_df=df.fillna({
    "temperature":0,
    "windspeed":0,
    "event":"No Event"
})
new_df
```

```python
new_df
```

**Use method-->Forwardfill (ffill) to determine how to fill na values**

```python
new_df=df.fillna(method="ffill")
new_df
```

**Use method-->Backwardfill (bfill) to determine how to fill na values**

```python
new_df=df.fillna(method='bfill')
new_df
```

**Use of axis**

```python
new_df=df.fillna(method='bfill',axis='columns')
new_df
```

**Limit parameter**

```python
new_df=df.fillna(method='bfill',limit=1)  #keeping limit it will only copy the calue once
new_df
```

**Interpolate**

```python
new_df=df.interpolate()  #Will give values in between
new_df
```

```python
new_df = df.interpolate(method="time") #Will give importance to time lean to values based on time
new_df
```

**Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)**

There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')   #how parameter indicates to drop onyl rows that are entirely Nan
new_df
```

```python
new_df = df.dropna(thresh=1)  #tresh parameter indicates to keep rows with atleast one valid value.
new_df
```
