**Handling Missing Data - Replace method**.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Dell\Downloads\weather_data (2).csv")
```

```python
df
```

**Replacing single value**

```python
new_df=df.replace(-99999,np.nan)   #If there was some other value like -88888 you could pass it as list like [-99999,-88888]
new_df
```

**Replacing per column**

```python
new_df=df.replace({"temperature":-99999,
                  "windspeed":-99999,
                  "event":'0'},np.nan)
new_df
```

**Regex**

```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df
```

**Replacing list with another list**

```python
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
df=df.replace(['exceptional','average','good','poor'],[10,7,8,5])
```
