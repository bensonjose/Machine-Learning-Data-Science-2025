**Read Excel**

```python
import pandas as pd
```

```python
df=pd.read_excel(r"C:\Users\Dell\Downloads\stock_data.xlsx")
df
```

```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Sam Walton'
    return cell

def convert_price_cell(cell):
    if cell=="n.a.":
        return 50
    return cell
    
df = pd.read_excel(r"C:\Users\Dell\Downloads\stock_data.xlsx","Sheet1", converters= {
        'people': convert_people_cell,
        'price': convert_price_cell
    })
df
```

**Write to Excel**

```python
df.to_excel("new.xlsx", sheet_name="stocks", index=False, startrow=2, startcol=1)
```
