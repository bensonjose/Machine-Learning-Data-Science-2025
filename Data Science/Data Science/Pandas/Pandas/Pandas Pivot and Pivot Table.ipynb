**Pivot and Pivot Table**

**<h1>Pivot Basics</h1>**

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Dell\Downloads\weather.csv")
```

```python
df
```

```python
df.pivot(index='date',columns='city')
```

```python
df.pivot(index='date',columns='city',values='humidity')
```

```python
df.pivot(index='humidity',columns='city')
```

**<h1>Pivot Table</h1>**

```python
df=pd.read_csv(r"C:\Users\Dell\Downloads\weather2.csv")
```

```python
df
```

```python
df.pivot_table(index='city',columns='date')
```

```python
df.pivot_table(index="city",columns="date", margins=True,aggfunc=np.sum)
```

<h1>Margins</h1

**Margins add a new column ie All---> Basically gives you col row aggregate**

```python
df.pivot_table(index='city',columns='date',margins=True,aggfunc=np.mean)   
```
